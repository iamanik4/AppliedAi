```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from sklearn.linear_model import SGDRegressor
```

```python
import numpy as np
import scipy as sp
import scipy.optimize

def angles_in_ellipse(num,a,b):
    assert(num > 0)
    assert(a < b)
    angles = 2 * np.pi * np.arange(num) / num
    if a != b:
        e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
        tot_size = sp.special.ellipeinc(2.0 * np.pi, e)
        arc_size = tot_size / num
        arcs = np.arange(num) * arc_size
        res = sp.optimize.root(
            lambda x: (sp.special.ellipeinc(x, e) - arcs), angles)
        angles = res.x 
    return angles
```

```python
a = 2
b = 9
n = 50

phi = angles_in_ellipse(n, a, b)
e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
arcs = sp.special.ellipeinc(phi, e)

fig = plt.figure()
ax = fig.gca()
ax.axes.set_aspect('equal')
ax.scatter(b * np.sin(phi), a * np.cos(phi))
plt.show()
```

```python
alpha_lst = [0.0001,1,100]
outlier = [(0,2),(21, 13), (-23, -15), (22,14), (23, 14)]
for i in range(len(alpha_lst)):
    plt.figure(figsize = (20,20))
    k = 0
    X= b * np.sin(phi)
    Y= a * np.cos(phi)
    for j in outlier:
        plt.subplot(3,5,k+1)
        k+=1 
        X = np.append(X,j[0]).reshape(-1,1)
        Y = np.append(Y,j[1]).reshape(-1,1)
        clf = SGDRegressor(alpha=alpha_lst[i], eta0=0.001, learning_rate='constant',random_state=0)
        clf.fit(X,Y)
        plt.scatter(X,Y,color='blue')
        plt.plot(X,clf.predict(X),color='red')
        plt.title('Outlier vs Position of hyper plane')
        
    
    plt.show()
        
```

Oberservation:
1. In first column of the plot when no outlier is present, hyperplane is perfectly bisecting the points
2. When only 2 outlier is introduced then we can see for lower values of alpha,hyperplane tends to deviate from the correct direction towards outlier. But when alpha is 100 we can observe effect of outlier is almost negligible.
3. In the 3rd column when three outlier were introduced hyper plane is more align in the directions of outlier. Even with high value of alpha significent effect is pronounced to the direction of hyper plane.
4. when number of outliers is increasing hyperplane tends to move towards outlier.

Conclusion:
More the number of outliers more will the change in the direction of hyperplane towards outliers.
Higher aplha can reduce the impact of outlier to some extent.
Magnitude of outlier when compare to datapoint also effect the position such that how much change will be pronounced.
